```python
import pandas as pd

df1 = pd.read_csv('query (4).csv')
df2 = pd.read_csv('sparql_2026-04-20_13-37-41Z.csv')

merged = pd.merge(df1, df2, left_on='cityName', right_on='cityName', how='inner')

merged.to_csv('Geopolitik Dataset - Graf.csv', index=False)
```
